### concat() - 행 방향 연결(아래로 쌓기)

```python
#데이터 생성: 1등석/2등석/3등석 승객을 각각 따로 분리
#'pclass':1등석 승객의 passenger_id를 tickets에서 필터링
p1 = passengers[passengers['passenger_id'].isin(tickets[tickets['pclass']==1]['passenger_id'])]
p1
```

```python
p2 = passengers[passengers['passenger_id'].isin(tickets[tickets['pclass']==2]['passenger_id'])]
p2
```

```python
p3 = passengers[passengers['passenger_id'].isin(tickets[tickets['pclass']==3]['passenger_id'])]
p3
```

```python
#기본 concat: 여러 DataFrame을 행 방향으로 단순 연결
result_raw = pd.concat([p1,p2,p3])
print("기본 concat (인덱스 중복):")
print(result_raw.index.tolist())
result_raw
```

```python
#연결 수 인덱스를 0부터 새로 부여
result_reset = pd.concat([p1,p2,p3],ignore_index=True)
print('ignore_index=True (인덱스 재설정):')
print(result_reset.index.tolist())
result_reset
```

### concat() - join 옵션(열 이름이 다를 때)

```python
#데이터 생성
df_a = pd.DataFrame({
    'passenger':[1,2,3],
    'name':['Alice','Bob','Carol'],
    'age':[29,35,24]
})
df_a
```

```python
df_b = pd.DataFrame({
    'passenger':[4,5,6],
    'name':['Dave','Eve','Frank'],
    'ticket_no':['A123','B456','C789']
})
df_b
```

```python
#열 이름이 일부 다른 두 DataFrame을 concat할 때 열 불일치 처리 방법
#outer join (기본 값 join='outer')
result_outer = pd.concat([df_a,df_b],ignore_index=True)
print('outer join(기본):')
print(result_outer)

#inner join: 양쪽에 공통으로 있는 열만 유지
result_inner = pd.concat([df_a,df_b],join='inner',ignore_index=True)
print("inner join(공통 열만):")
print(result_inner)
```

### merge() - left/right/outer join 비교

```python
#일부 승객은 티켓 정보가 없는 상황으로 비교
passengers_full = passengers.merge(tickets, on='passenger_id', how='left')
print('left join 결과:')
display(passengers_full)
```

```python
tickets_partial = tickets[tickets['passenger_id']<=800].copy()
tickets_partial
```

```python
#left join:왼쪽(passengers) 전체 유지, 오른쪽(tickets)에 없으면 NaN
left_result = passengers_full.merge(tickets_partial, on='passenger_id',how='left', suffixes=('_x', '_y')) # 중복 컬럼을 명확히 하기 위해 suffixes 명시
print(f"left join:{len(left_result)}행 (승객 전체 유지)")
print(f"fare_y NaN 수:{left_result['fare_y'].isnull().sum()}") # 'fare' 대신 'fare_y' 사용
display(left_result)

#right join: 오른쪽(tickets) 전체 유지, 왼쪽(passengers)에 없으면 NaN
right_result = passengers_full.merge(tickets_partial, on='passenger_id', how='right', suffixes=('_x', '_y'))
print(f"right join:{len(right_result)}행 (티켓 전체 유지)")
display(right_result)

#outer join: 양쪽 전체 유지, 없는 쪽은 NaN
outer_result = passengers_full.merge(tickets_partial, on='passenger_id', how='outer', suffixes=('_x', '_y'))
print(f"outer join:{len(outer_result)}행 (전체 유지)")
display(outer_result)
```

### 데이터 정규화 - 중복 제거 및 테이블 분리

```python
#정규화 전: 중복이 많은 단일 테이블(같은 등급 승객마다 등급 정보가 반복)
wide_table = passengers.merge(tickets, on='passenger_id').merge(class_info, on='pclass')
print(f"정규화 전 테이블:{wide_table.shape}")
print(f"'class_name' 중복 수:{wide_table['class_name'].duplicated().sum()}") #중복 개수 집계
```

```python
wide_table
```

```python
#등급 정보 테이블 분리
class_table = wide_table[['pclass','class_name','class_desc','base_fare']].drop_duplicates().reset_index(drop=True)
print(f"[분리] 등급 정보 테이블:{class_table.shape}") #중복 행 제거
print(class_table)

#메인 테이블에서 분리된 열 제거
main_table = wide_table.drop(columns=['class_name','class_desc','base_fare'])
print(f"[정리] 메인 테이블:{main_table.shape}")
print(main_table.head(3))

#필요할 때 merge로 복원
restored = main_table.merge(class_table, on='pclass', how='left')
print(f"[복원] 병합 후:{restored.shape}")
print(restored.head(3))

#원본과 동일한지 검증
assert set(restored.columns) == set(wide_table.columns), '열 불일치'
assert len(restored) == len(wide_table), '행 수 불일치'
print('정규화 후 복원 검증 통과')

#정규화 효과 비교
before_size = wide_table[['pclass','class_name','class_desc','base_fare']].memory_usage(deep=True).sum()
after_size = class_table.memory_usage(deep=True).sum() #deep=True: 문자열 실제 메모리 포함
print(f"정규화 전 등급 정보 메모리:{before_size:,}bytes")
print(f"정규롸 후 등급 정보 메모리:{after_size:,}bytes")
print(f"절감율:{(1-after_size/before_size)*100:.1f}%")
```

### 종합 - 타이타닉 전체 전처리 파이프라인 최종 완성

```python
titanic_final = pd.read_csv('/content/drive/MyDrive/KWU/bigdata/dataPreProcessing/train.csv')
titanic_final = titanic_final.drop(columns=['Cabin'])

#데이터 결합
titanic_final = titanic_final.merge(port_info, left_on='Embarked',right_on='embarked',how='left')

# 결측치 처리
titanic_final['Age'] = titanic_final.groupby(['Sex','Pclass'])['Age'].transform(lambda x:x.fillna(x.median()))
titanic_final['Embarked'] = titanic_final['Embarked'].fillna(titanic_final['Embarked'].mode()[0])
titanic_final['port_name'] = titanic_final['port_name'].fillna('Southampton')
titanic_final['country'] = titanic_final['country'].fillna('영국')

# 이상치 처리 - 캡핑
Q1, Q3 = titanic_final['Fare'].quantile([0.25,0.75])
titanic_final['Fare'] = titanic_final['Fare'].clip(upper=Q3 + 1.5 * (Q3 - Q1)) #초과 값을 상한으로 대체

# 분포 변환 - 로그 변환
titanic_final['Fare'] = np.log1p(titanic_final['Fare']) #양의 왜도

# 파생 변수 생성
titanic_final['FamilySize'] = titanic_final['SibSp'] + titanic_final['Parch'] + 1
titanic_final['IsAlone'] = (titanic_final['FamilySize'] == 1).astype(int)
titanic_final['Title'] = titanic_final['Name'].str.extract(r',\s*([^\,]+)\.').squeeze()
titanic_final['Title'] = titanic_final['Title'].apply(lambda x:x if x in ['Mr','Miss','Mrs','Master'] else 'Other')

# 범주형 인코딩
# 명목형 변수(순서 없음) -> 원-핫 인코딩
titanic_final = pd.get_dummies(titanic_final, columns=['Sex','Embarked','Title'],drop_first=True)

# 순서형 변수(순서 있음) -> 순서형 인코딩
titanic_final['Pclass'] = titanic_final['Pclass'].map({1:2,2:1,3:0})

# 불필요 열 제거
titanic_final = titanic_final.drop(columns=['Name','Ticket','SibSp','Parch','port_name','country','embarked'])

# 스케일링
from sklearn.preprocessing import StandardScaler #평균=0, 표준편차=1로 표준화
num_cols = ['Age','Fare','FamilySize']
scaler = StandardScaler()
titanic_final[num_cols] = scaler.fit_transform(titanic_final[num_cols])

# 최종 결과 확인
print(f"\n최종 shape:{titanic_final.shape}")
print(f"결측치:{titanic_final.isnull().sum().sum()}개")
print(f"\n최종 컬럼 목록:")
print(titanic_final.columns.tolist())
print(f"\n데이터 미리보기:")
print(titanic_final.head(3).round(3))
```
