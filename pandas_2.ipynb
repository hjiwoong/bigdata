### 시리즈와 데이터프레임 데이터 변경하기

### 열 내용 변경하기

```python
print(scientists['Age'])
```

```python
print(scientists['Age'].sample(frac=1,random_state=42))
#frac=1 , frac인자는 샘플링할 비율을 지정, 전체데이터(100%)를 샘플링하겠다는 의미, 모든 데이터를 포함하되 순서만 무작위로 섞는 셔플링 효과를 준다.
#random_state 인자는 난수 생성기의 시드(seed)를 설정, 이 값을 지정하면 코드를 여러번 실행해도 항상 동일한 순서로 데이터가 섞이게 된다. 결과의 재현성을 보장
```

```python
scientists["Age"]=scientists["Age"].sample(frac=1,random_state=42) #섞기 전과 동일
print(scientists["Age"])
```

```python
scientists['Age'] = scientists['Age'].sample(frac=1, random_state=42).values
print(scientists['Age'])
```

```python
scientists['age_days']=(scientists['died_dt']-scientists['born_dt'])
display(scientists)
```

### 데이터 프레임에서 행 삭제하기

```python
scientists=scientists.set_index('Name') #Name 컴럼 값으로 데이터 프레임의 행 인덱스 설정
```

```python
display(scientists.head())
```

```python
scientists_without_franklin=scientists.drop(index=['Rosaline Franklin'])
```

```python
scientists_without_franklin.head() #한 행이 삭제된 데이터프레임
```

```python
scientists_without_multiple=scientists.drop(index=['Rosaline Franklin','William Gosset'])
```

```python
display(scientists_without_multiple) #여러 행이 삭제된 데이터프레임
```

```python
display(scientists.head()) #원본은 유지하고 새로운 변수를 이용해 내용 삭제
```

```python
scientists['age_years']=(scientists['age_days'].dt.days/365).apply(np.floor)
#.dt는 시리즈의 날짜/시간 관련 속성에 접근하게 해준다
#.days는 day정보만 정수로 추출한다
#.dt.day로 숫자로 변경
display(scientists.head())
```

```python
#assign()으로 열 수정하기
scientists=scientists.assign(
    age_days_assign=scientists['died_dt']-scientists['born_dt'],
    age_years_assign=(scientists['age_days'].dt.days/365).apply(np.floor)
)
display(scientists.head())
#apply() 이 시리즈의 각 값에 대해 특정 함수를 적용하고 싶을 때 사용
#assign() 기존 데이터프레임을 변경하지 않고 새로운 열이 추가되거나 변경된 새로운 데이터프레임을 반환
#괄호 안에는 새로 생성할 열의 이름과 그 값을 지정
#열이름 = 값
```

```python
scientists=scientists.assign(
    age_days_assign=scientists['died_dt']-scientists['born_dt'],
    age_years_assign=lambda df_:(df_['age_days_assign'].dt.days/365).apply(np.floor)
)
display(scientists.head())
#df_라는 인자로 현재 assign()이 적용되고 있는 데이터프레임(즉, scientists 데이터프레임)을 참조할 수 있다.
```
