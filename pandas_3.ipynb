### pickle 파일 저장하고 불러오기
- 파이썬 프로그램에서 생성한 어떤 객체(예: 리스트, 딕셔너리, pandas DataFrame, Series등)의 구조와 데이터를 그대로 파일로 저장하고 싶을 때 pickle모듈을 사용

- 머신러닝 모델 내부의 가중치와 설정값이 그대로 유지

### JSON 파일로 저장하기
```
#객체(딕셔너리와 유사) - 중괄호 사용, 키와 값은 큰 따옴표로 감싸기
{"title":"어벤져스","year":2012}

#배열(리스트와 유사) - 대괄호 사용
{"director":["조스 웨던","앤서니 루소"]}

#중첩 구조 - 여러 객체를 배열로 묶기
[
  {"title":"어벤져스", "year":2012},
  {"title":"어벤져스:엔드게임", "year":2019}
]
```

````python
import json
````

````python
d={"title":"어벤져스", "year":2012}
#딕셔너리 -> JSON 문자열로 변환
d_str = json.dumps(d)
type(d_str)
````

````python
print(d_str)
````

````python
#JSON문자열 -> 파이썬 딕셔너리로 변환
d2 = json.loads(d_str)
print(d2)
````

````python
print(d2['title'])
````

````python
#세겹 따옴표(""")로 여러 줄 문자열 작성
d3_str = """
{
  "title":"어벤져스",
  "director":["조스 웨던"],
  "year":2012
}
"""
d3 = json.loads(d3_str) #JSON 문자열 -> 파이썬 딕셔너리로 변환
````

````python
d3
````

````python
#JSON 문자열(배열구조)을 데이터프레임으로 변환
df=pd.read_json(d3_str)
df
````

````python
df = pd.DataFrame(json.loads(d3_str)) #또는 파이썬 딕셔너리 객체를 DataFrame 클래스로 변환
df
````

### JSON으로 저장하기

````python
sci_json=sci_sub_dict.to_json(orient='records',indent=2,date_format="iso")
pprint.pprint(sci_json)
#pandas 데이터프레임을 json문자열로 변환
#orient='records'는 데이터프레임의 각 행을 하나의 json객체로 변환하고 리스트에 담는 형식
#indent는 들여쓰기
#date_format는 날짜/시간 정보, "iso"는 표준화된 방식으로 표현
````
