```python
nums = []
for i in range(100):
    try:
        div = i - 50
        nums += [100/div]
    except Exception as e:
        print(e)
        print(i, div)
```

```python
num = [1,20,300]
try:
    first = nums[3]/1
except IndexError:
    print("인덱스 범위를 벗어났습니다")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
print("프로그램이 정삭적으로 종료되었습니다")
```
