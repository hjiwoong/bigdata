```python
#필요한 라이브러리 임포트
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
#iris 데이터셋 불러오기
iris = load_iris()
```

```python
iris
```

### pandas dataFrame

```python
#데이터프레임으로 변환
df = pd.DataFrame(iris.data,columns=['sepal_length','sepal_width','petal_length','petal_width'])
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
#데이터 확인
print("=== 데이터 상위 5행 ===")
print(df.head())
```

```python
print("\n=== 데이터 형태 (행, 열) ===")
print(df.shape)
```

```python
print("\n=== 칼럼 유형 확인 ===")
print(df.dtypes)
```

```python
print("\n===결측치 확인 ===")
print(df.isnull().sum())
```

```python
print("\n=== 종류별 데이터 개수 ===")
print(df['species'].value_counts())
```

###평균(Mean)

```python
#단일 칼럼 평균
print("꽃받침 길이 평균: ",df['sepal_length'].mean())
print(f"꽃받침 길이 평균: {df['sepal_length'].mean():.2f}")
```

```python
#전체 숫자형 칼럼 평균
print("전체 칼럼 평균:")
print(df.mean(numeric_only=True).round(2))
#(numeric_only=True) 숫자형 칼럼(정수, 실수)만 계산
```

```python
#종류별 평균 비교
print("종류별 꽃잎 길이 평균:")
print(df.groupby('species')['petal_length'].mean())
```

### 중앙값(Median)

```python
#단일 칼럼 중앙값
print("꽃받침 길이 중앙값:",df['sepal_length'].median())
```

```python
#전체 칼럼 중앙값
print("전체 칼럼 중앙값:")
print(df.median(numeric_only=True))
```

```python
#평균 vs 중앙값 비교
print("=== 평균 vs 중앙값 비교 ===")
comparison = pd.DataFrame({
    '평균': df.mean(numeric_only=True).round(2),
    '중앙값': df.median(numeric_only=True)
})
print(comparison)
```

### 최소값/최대값 (Min/Max)

```python
#최소값
print("=== 최소값 ===")
print(df.min(numeric_only=True))
```

```python
#최대값
print("=== 최대값 ===")
print(df.max(numeric_only=True))
```

```python
#범위(최대값 - 최소값)
print("=== 범위(Range) ===")
numeric_cols = df.select_dtypes(include='number')
print(numeric_cols.max() - numeric_cols.min())
#df에서 숫자형 데이터를 포함하는 칼럼만 선택하여 새로운 DataFrame인 numeric_cols를 생성
```

```python
#어떤 행이 최대값/최소값인지 확인
print("꽃받침 길이가 가장 긴 꽃:")
print(df['sepal_length'].idxmax())
print(df.loc[131]) #131행의 컬럼과 값이 출력
print("\n")
print(df.loc[df['sepal_length'].idxmax()])
#.idxmax(): 'sepal_length' 칼럼에서 가장 큰 값(최대값)을 가지는 행의 인덱스를 반환
#.loc[]: 인덱서를 사용하여 특정 행 또는 칼럼을 레이블(이름)을 기반으로 선택
```

### 분위수(Quantile/Percentile)

```python
#분위수 구하기
print("=== 분위수(25%, 50%, 75%) ===")
print(df.quantile([0.25,0.5,0.75],numeric_only=True))
```

```python
#특정 칼럼의 분위수
col = 'sepal_length'
Q1 = df[col].quantile(0.25)
Q2 = df[col].quantile(0.50)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
print(f"꽃받침 길이 분위수: ")
print(f" Q1 (25%): {Q1}")
print(f" Q2 (50%): {Q2}")
print(f" Q3 (75%): {Q3}")
print(f" IQR: {IQR}")
```

```python
#이상치 경계값
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print("이상치 경계값:")
print(f" 하한: {lower:.2f}")
print(f" 상한: {upper:.2f}")
```

```python
#이상치 확인
outliers = df[(df[col] < lower) | (df[col] > upper)]
print(f"이상치 데이터 수: {len(outliers)}개")
#df[True, False, True] DataFrame에 불리언 시리즈를 전달하면 True로 표시된 행들만 필터링하여 새로운 DataFrame인 outliers을 반환
```

### 분산(Variance)

```python
#단일 칼럼 분산
print("꽃받침 길이 분산:", df['sepal_length'].var())
print("\n")
print(f"꽃받침 길이 분산: {df['sepal_length'].var():.2f}")
```

```python
#전체 칼럼 분산
print("=== 전체 칼럼 분산 ===")
print(df.var(numeric_only=True).round(2))
```

```python
#종류별 분산 비교(분산이 크면 데이터가 넓게 퍼져있음)
print("종류별 꽃잎 길이 분산")
print(df.groupby('species')['petal_length'].var().round(2))
```

### 표준 편차 (Standard Deviation)

```python
#단일 칼럼 표준편차
print("꽃받침 길이 표준편차:",df['sepal_length'].std())
print("\n")
print(f"꽃받침 길이 표준편차: {df['sepal_length'].std():.2f}")
```

```python
#전체 칼럼 표준편차
print("=== 전체 칼럼 표준편차 ===")
print(df.std(numeric_only=True).round(2))
```

```python
#분산 vs 표준편차 비교정리
print("=== 분산 vs 표준편차 비교 ===")
summary = pd.DataFrame({
    '분산': df.var(numeric_only=True).round(4),
    '표준편차': df.std(numeric_only=True).round(4)
})
print(summary)
#표준편차(Standard Deviation): 분산의 제곱근으로 데이터가 평균으로부터 평균적으로 얼마나 떨어져있는지를 나타낸다.
#데이터의 원래 단위를 따르기 때문에 분산보다 직관적으로 이해하기 쉽다

#petal_length(꽃잎 길이): 평균으로부터 가장 넓게 퍼져있으며 데이터 포인트들이 서로 크게 다르다는 것을 의미
#sepal_width(꽃받침 너비): 데이터가 평균 주변에 가장 밀집되어 있으며 데이터 포인트들이 서로 유사하다는 것을 의미
```

```python
#평균 ±표준편차 범위 (데이터의 약 68%가 이 범위에 속함)
col = 'sepal_length'
mean = df[col].mean()
std = df[col].std()
print("꽃받침 길이: 평균 ± 표준편차")
print(f" 평균: {mean:.2f}")
print(f" 표준편차: {std:.2f}")
print(f" 범위: {mean - std:.2f} ~ {mean + std:2f}")
count_in_range = df[(df[col] >= mean - std) & (df[col] <= mean + std)].shape[0]
#True로 표시된 행들만 필터링하여 새로운 DaraFrame인 count_in_range을 반환
#.shape[0]: 필터링된 DataFrame의 행의 개수를 반환
print(f"이 범위에 속하는 데이터: {count_in_range}개 ({count_in_range/len(df)*100:.1f}%)")
```

### 데이터프레임에서 기술통계 한번에 구하기

```python
#기본 기술 통계(숫자형 칼럼 자동 선택)
print("=== 기술통계 요약(describe) ===")
print(df.describe().round(2))
```

```python
#범주형 포함 전체 기술 통계
print("=== 범주형 포함 기술통계 ===")
print(df.describe(include='all').round(2))
#NaN은 'Not a Number'의 줄임말, '결측값' 또는 '누락된 값'을 나타내는 데 사용
```

```python
#종류별 기술 통계 (groupby + describe)
print("=== 종류별 기술통계 ===")
print(df.groupby('species').describe())
```

```python
#내가 원하는 통계만 선택해서 보기
print("=== 커스텀 기술 통계 ===")
custom_stats = df.groupby('species')['petal_length'].agg([
    ('평균', 'mean'),
    ('중앙', 'median'),
    ('표준편차', 'std'),
    ('최소값', 'min'),
    ('최대값', 'max')
]).round(3)
print(custom_stats)
#.add([...]): 하나 이상의 집계(aggregation) 함수를 적용하는 메서드
#('평균', 'mean'): 'mean'함수를 적용하여 평균을 계산하고, 결과 칼럼의 이름을 '평균'
```
