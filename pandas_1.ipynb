```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":["kim","lee","park","choi","na"],
                   "test": [45, 30, 40, 37, 48],
                   "kor": [20,17,22,18,24],
                   "eng": [19,14,18,15,25]})
```

```python
display(df)
print(df)
```

```python
df = pd.DataFrame([["kim",45,20,19],
                   ["lee",30,17,14],
                   ["park",40,22,18],
                   ["choi",37,18,15],
                   ["na",48,24,25]],
                  columns=["name","test","kor","eng"])
display(df)
```

```python
df=pd.DataFrame()
df["name"]=["kim","lee","park","choi","na"]
df["test"]=[45,30,40,37,48]
df["kor"]=[20,17,22,18,24]
df["eng"]=[19,14,18,15,25]
df["math"]=[10,20,30,40,50]
display(df)
```

```python
display(df["math"]) #인덱스 번호와 값만 존재하는 1차원 배열 "시리즈"
```

```python
display(df[["math","name"]]) #2차원 배열 "데이터프레임"은 대괄호 두개 사용
```

```python
n=df.pop("math")
```

```python
display(n)
display(df)
```

```python
for index, row in df.iterrows():
  print(index, row["name"], row["test"])
```

```python
df["sum"]=df["test"]+df["kor"]+df["eng"]
display(df)
```

```python
display(df[df["sum"]>=80]) #조건을 만족하는 데이터 행을 추출
```

```python
print(df["sum"]>=80)
```

```python
display(df[df["sum"]>=80][["name", "sum"]])
```
