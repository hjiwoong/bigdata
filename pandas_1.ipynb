```python
#기초통계
display(df)
```

```python
display(df.groupby('year')['lifeExp'].mean())
```

```python
#2개 이상 열 그룹화 하기
multi_group_var = df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
display(multi_group_var)
#year continent는 멀티 인덱스
```

```python
flat=multi_group_var.reset_index() #기존 멀티 인덱스였던 year.continent가 일반 열로 바뀌고 새로운 0부터 시작하는 인덱스 생성
display(flat)
```

```python
#그룹화한 데이터 개수 구하기
print(df.groupby('continent')['country'].nunique())
#Series나 DataFrame에서 고유한(unique) 값들의 개수를 반환. 각 대륙별로 고유한 국가의 개수를 계산
```

```python
print(df.groupby('continent')['country'].value_counts())
#선택된 'country'열에서 각 대륙별로 각 국가가 몇번 나타나는지를 계산
#Series 형태로 반환하고 가장 빈도가 높은 값부터 내림차순으로 정렬
#Series 내의 고유한 값들과 각 값의 출현 빈도를 계산
```
