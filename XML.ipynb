```python
#XML
import xml.etree.ElementTree as ET #XML 데이터를 처리하기 위한 ElementTree 모듈

# XML 문자열
x_str = """
  <movie>
    <title>어벤져스</title>
    <director>조스 웨던</director>
    <year>2012</year>
  </movie>
"""
#XML문자열 -> XML 객체로 변환
movie = ET.fromstring(x_str) #XML Element 객체로 파싱
type(movie)
```

```python
print(movie.tag)
```

```python
movie
```

```python
#자식 엘리먼트를 리스트로 변환
children = list(movie)
children
```

```python
#findtext()로 특정 태그의 텍스트 추출
print(movie.findtext("title"))
print(movie.findtext("director"))
print(movie.findtext("year"))
```

```python
#JSON
import json #JSON 데이터를 처리하기 위한 json 모듈
d={"title":"어벤져스","year":"2012"} #파이썬 딕셔너리
#딕셔너리 -> JSON 문자열로 변환
d_str = json.dumps(d) #JSON 형식의 문자열로 변환
type(d_str)
```

```python
d_str
```

```python
#JSON 문자열 -> 파이썬 딕셔너리로 변환
d2 = json.loads(d_str)
print(type(d2))
```

```python
print(d2['title'])
```
